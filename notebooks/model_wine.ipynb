## Passo 1 - Setup do projeto

```python
import joblib
import os
import urllib

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

print('Matplot version: {}'.format(matplotlib.__version__))
print('Numpy version: {}'.format(np.__version__))
print('Pandas version: {}'.format(pd.__version__))
print('Scikit-learn version: {}'.format(sklearn.__version__))
print('Seaborn version: {}'.format(sns.__version__))

WINE_ROOT = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/'
WINE_URL = WINE_ROOT + 'wine.data' 

DATA_PATH = os.path.join('..', 'data', 'raw')
DATA_FILE = 'wine.csv'
WINE_DATA = os.path.join(DATA_PATH, DATA_FILE)

DEPLOY_PATH = os.path.join('..', 'model')
DEPLOY_FILE = 'finalized_model.sav'

RANDOM_STATE = 42

print('\nConfiguração completa !')
```

# Passo 2 - Recuperação dos Dados

```python
def download_data(data_url, data_path, data_file):
    os.makedirs(data_path, exist_ok=True)
    local_path = os.path.join(data_path, data_file)
    urllib.request.urlretrieve(data_url, local_path)
```

```python
download_data(WINE_URL, DATA_PATH, DATA_FILE)
```

```python
col_names = ['class', 'alcohol', 'malic_acid', 'total_phenols', 'flavanoids', 'nonflav_phenols',
             'proanthocyanins', 'color']
```

```python
def load_data(data_path, data_file):
    local_path = os.path.join(data_path, data_file)
    return pd.read_csv(local_path, header=None, usecols=[0,1,2,6,7,8,9,10], names=col_names)
```

```python
wine_df = load_data(DATA_PATH, DATA_FILE)
```

## Correlação e Distribuição

```python
x = wine_df[['alcohol', 'malic_acid', 'total_phenols', 'flavanoids', 'nonflav_phenols', 'proanthocyanins', 'color']]
y = wine_df['class']
```

```python
cores = y.astype('category').cat.codes
_ = pd.plotting.scatter_matrix(x, c=cores, figsize=[18,18], marker='D')
```

# Passo 4 - Treinamento do Modelo

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.15, stratify=y, random_state=42)
```

```python
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy  = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i]  = knn.score(x_test, y_test)
```

```python
plt.title('k-NN: Acuracia por numero de vizinhos')
plt.plot(neighbors, train_accuracy, label='Acuracia no treinamento')
plt.plot(neighbors, test_accuracy, label='Acuracia no teste')
plt.legend()
plt.xlabel('Numero de vizinhos')
plt.ylabel('Acuracia')
plt.show()
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train, y_train)
```

```python
predicoes = model.predict(x_test)
print('Predicoes: {}'.format(predicoes))
```

```python
acuracia = model.score(x_test, y_test)
print('Acuracia do modelo: {}'.format(acuracia))
```

```python
cm = confusion_matrix(y_test, predicoes)
plt.figure(figsize=(10,5))
_ = sns.heatmap(cm, annot=True, fmt='d', cbar=False)
```

```python
print('Precisão, revocação e F1-score:')
print(classification_report(y_test, predicoes))
```

# Passo 6 - Deploy do Modelo

```python
local_path = os.path.join(DEPLOY_PATH, DEPLOY_FILE)
joblib.dump(model, local_path)
```
